```python
# Data collected from various mobile and pc games also play a major role these days. We have a dataset containing Pokémon data of the characteristics of all major Pokémon’s. But the Pokémon’s are in need of some insights that they are unable to sort out. Help the below Pokémon’s to visually understand the data that they need.
    
#     i.	Ash wants to analyze attack and defending powers. Help him do it with an appropriate plot.
#     ii.	Brock wants to understand the Pokémon types and their attacks.Help him to visualize it with an appropriate  plot.
#     iii.	Misty, as part of her research needs to know the attack distribution. Help her with a proper distribution plot.
#     iv.	Oak after losing many games wanted to know the solidity relationship among attack and defense. Visualize density plot for oak to progress.
#     v.	Gary wants to know count of different types of Pokémon. Can you assist him with appropriate categorical distribution?


import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
df = pd.read_csv(r"C:\Users\Navaneeth Kumar ch\Downloads\Pokemon.csv")
df.head()
```

```python
box_df = df.drop(['Total', 'Legendary'], axis=1)
sns.boxplot(data=box_df)
```

```python
#ii.
sns.set_style('whitegrid')
sns.violinplot(x='Type 1', y='Attack', data=df)
```

```python
#iii. Distribution Plot (Histogram)
sns.histplot(df.Attack)
```

```python
#iv.Density Plot
sns.kdeplot( df.Defense)
```

```python
#iv.Density Plot
sns.kdeplot(df,x='Attack', y='Defense')
```

```python
#v.
sns.countplot(x='Type 1', data=df)
plt.xticks(rotation=-45)
```
