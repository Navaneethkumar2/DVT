```python
# 1.
# Now-a-days data comes with various dimensions and entities. To understand the data properly and draw conclusions it is essential to understand the flow among the data. Niharith works in an MNC Company where he was assigned a task to understand the flow of data and paths for the below two scenarios. As he had to present it to his colleagues in a conference, he thought visualizing this flow would be more effective to present as people understand visuals clearly than oral explanations. Help Niharith create an appropriate flow chart/diagram using python for them.

# 1.	Two rival companies named Dango and Wipro face huge losses and join hands with a new company. The three of them start a company name Imtac together. Any project sent to Imtac gets reviewed and tested in two phases. The first phase is Orizon, once the project passes the test at Orizon the second phase of testing is done at Eclipse. Niharith is supposed to explain their clients the above scenario.

import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

df = pd.DataFrame({ 'from':['Imtac', 'Wipro', 'Orizon','Imtac'], 'to':['Dango', 'Imtac', 'Eclipse','Orizon']})
G=nx.from_pandas_edgelist(df, 'from', 'to')
fig, ax= plt.subplots()
nx.draw(G,with_labels=True, ax=ax)
plt.show()
 
```

```python
# 2.
# Any project that is assigned to Nihariths company undergoes 3 phases of testing and then the final output of the project is submitted. MIM , HMC and ITC are the three companies that offer projects to Nihariths company. Niharith is also supposed to show the below data in the diagram.
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.sankey import Sankey
 
# basic sankey chart
Sankey(flows=[0.25, 0.15, 0.60, -0.15, -0.05, -0.50, -0.10], labels=['ITC', 'MIM', 'HMC', '1st Phase', '2nd Phase', 'Final', '3rd Phase'], orientations=[-1, 1, 0, 1, 1, 0,-1]).finish()
plt.title("Project Phases and Flow of the Project")
```
