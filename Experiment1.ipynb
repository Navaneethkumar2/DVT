```python
# You are the physical director at KLH and you are given the I.D. No’s of students who are interested in two different sports namely cricket and volleyball. KL Vijaywada is conducting an inter-college competition for cricket and volleyball and you should present the data to the higher authorities to show the no. of students interested in cricket only, volleyball only and both cricket and volleyball. 
# Prepare a proper chart and present it.


from matplotlib import pyplot as plt
%matplotlib inline
from matplotlib_venn import venn2
cricket=[1803,1804,1805,1810,2000,2010,2015,2016,1895,1998,2020]
volleyball=[2020,1803,1805,2222,3333,2016,2014,1889,1810,1996]
venn2([set(cricket), set(volleyball)],set_labels=('cricket', 'volleyball'))
plt.show()
```

```python
# Nihanth works as a data analyst in a forecasting company where in his colleagues asked him the details about no. of people who were affected by Ebola Virus from different areas. Nihanth has the data in a tabular form but he thought that giving them a chart showing no. of people affected would be easier to understand. Help Nihanth visualize the data he has using appropriate plotting technique.
# [40,35,20,88,250,74] citys=['Hyderabad','delhi','mumbai','chennai','Rajasthan','Bbngalore']

import pandas as pd
import squarify
numvirus=[40,35,20,88,250,74]
citys=['Hyderabad','delhi','mumbai','chennai','Rajasthan','Bngalore']
df = pd.DataFrame({'num_virus':numvirus, 'citys':citys })
squarify.plot(sizes=df['num_virus'], label=df['citys'], alpha=.8 )
plt.axis('off')
plt.show()
```

```python
# Case-Study -3 
# ou are appointed as the Vice President of Imtac Company. The staff in the company are divided into three departments namely ‘Advertising’, ‘IT’ and ‘Maintenance’. The coordinating team members id and expenses are given below. Using a donut chart depict the sub department’s expenses so that it can be presented to the President of the company in the next conference. 

import matplotlib.pyplot as plt
deparment=['advertising','IT','maintenance']
dpmembers=['a1','a2','a3','it1','it2','it3','m1','m2','m3']
dpmembers_exp=[100,300,522,650,250,100,600,450,300]
deparment_size=[922,1000,1350]
a, b, c=[plt.cm.Blues, plt.cm.Reds, plt.cm.Greens]
fig, ax = plt.subplots()
ax.axis('equal')
mypie, _ = ax.pie(deparment_size, radius=1.3, labels=deparment, colors=[a(0.6), b(0.6), c(0.6)] )
plt.setp( mypie, width=0.3, edgecolor='white')


mypie2, _ = ax.pie(dpmembers_exp, radius=1.3-0.3, labels=dpmembers, labeldistance=0.7, colors=[a(0.5), a(0.4), a(0.3), b(0.5), b(0.4), c(0.6), c(0.5), c(0.4), c(0.3), c(0.2)])
plt.setp( mypie2, width=0.4, edgecolor='white')
plt.margins(0,0)
 
# show it
plt.show()

```
