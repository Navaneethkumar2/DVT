```python
import pandas as pd
import xml.etree.ElementTree as et 
import seaborn as sns
```

```python
# Ahswin is the Head Boy of his school. He is given the data of students' details like name, mail, grade, age and no. of hours they studied and was given the task of finding out if there is any relationship between grades that the students obtained and the no. of hours they studied. He was also asked to show the age distribution and how spread extent of grade over student. The difficulty he faced is that he was given an xml file. So, help Ashwin do his task using python.
xtree = et.parse(r"C:\Users\Navaneeth Kumar ch\Downloads\student.txt")
xroot = xtree.getroot()
dic={'name':[], 'email': [],'grade': [], 'age': [],'hrs': []}
for node in xroot: 
    dic['name']. append(node.attrib.get("name"))
    dic['grade'].append(node.find("grade").text)
    dic['age'].append(node.find("age").text)
    dic['hrs'].append(node.find('hrs').text)
df = pd.DataFrame.from_dict(dic) 
df['hrs']=df['hrs'].astype(int)
df['grade']=df['grade'].astype(float)
df['age']=df['age'].astype(int)
df['email']=df['email'].astype(str)
df['name']=df['name'].astype(str)
```

```python
df['age'].plot(kind='hist',bins=7)
```

```python
sns.regplot(x="hrs", y="grade", data=df)
```
